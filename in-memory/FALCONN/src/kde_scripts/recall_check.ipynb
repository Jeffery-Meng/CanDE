```python
def avg_recall(gt, res):
    qn = gt.shape[0]
    recalls = np.zeros(qn)
    for i in range(qn):
        gnd = set(gt[i])
        result = set(res[i])
        recalls[i] = len(gnd.intersection(result)) / len(gnd)
    return recalls
```
